```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url)

parser = BeautifulSoup(response.content, 'html.parser')
jobs = parser.find_all('div', class_='card-content')

with open('jobs.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Company', 'Location', 'job detail url'])
    for card in jobs:
        title = card.find('h2', class_='title is-5').text
        company = card.find('h3', class_='subtitle is-6 company').text
        location = card.find('p', class_='location').text

        links = card.find_all('a', class_='card-footer-item')
        job_detail_url = links[1]('href')

        writer.writerow([title, company, location, job_detail_url])

print('scraping done! open file jobs.csv')
```
